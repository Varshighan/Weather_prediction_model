Z-score testing

```python
from scipy.stats import zscore

numeric_cols = df.select_dtypes(include='number').columns
z_scores = df[numeric_cols].apply(zscore)

outlier_counts = (abs(z_scores) > 3).sum()
print(outlier_counts)
```

| `wind_speed_10m (km/h)`    | 25               | Slight noise            | ⚠️ Optionally clip             |
| `wind_speed_100m (km/h)`   | 3                | Looks fine              | ✅ Leave                        |
| `wind_direction_*`         | 0                | Noisy but natural       | ✅ Leave as is                  |

Cap (clip) Outliers – Only for selected features

```python
def clip_outliers(df, cols, z_thresh=3):
    from scipy.stats import zscore
    for col in cols:
        z = zscore(df[col])
        df[col] = df[col].where(abs(z) < z_thresh, df[col].median())
    return df

outlier_cols = ['temperature_2m (°C)', 'surface_pressure (hPa)', 
                'wind_speed_10m (km/h)', 'wind_speed_100m (km/h)']

df = clip_outliers(df, outlier_cols)
```

```python
from scipy.stats import zscore

numeric_cols = df.select_dtypes(include='number').columns
z_scores = df[numeric_cols].apply(zscore)

outlier_counts = (abs(z_scores) > 3).sum()
print(outlier_counts)
```

```python
import os

header_lines = []
with open('Dataset/Tamil Nadu/Metrological Data/Coimbatore.csv', 'r') as f:
    for _ in range(3):
        header_lines.append(next(f))

output_dir = 'Dataset_cleaned/Tamil Nadu/Metrological Data/'
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, 'Coimbatore.csv')

df.to_csv(output_path, index=False)

with open(output_path, 'r') as original:
    cleaned_data = original.read()

with open(output_path, 'w') as final_file:
    final_file.writelines(header_lines)
    final_file.write(cleaned_data)
```
